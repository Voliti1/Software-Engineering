```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"안녕하세요, 저는 {self.name}이고, {self.age}살입니다.")
    
    def birthday(self):
        self.age += 1
        print(f"생일 축하합니다! {self.name}의 나이가 {self.age}살이 되었습니다.")

    def change_info(self, new_name, new_age):
        self.name = new_name
        self.age = new_age
        print(f"이름이 {self.name}로 나이가 {self.age}로 변경되었습니다")

class Worker(Person):
    def __init__(self, name, age, job):
        super().__init__(name, age)
        self.job = job

    def introduce(self):
        super().introduce()
        print(f"저는 {self.job}로 일하고 있습니다")

    def change_job(self, new_job):
        self.job = new_job
        print(f"직업이 {self.job}로 변경되었습니다")

    def work(self):
        print(f"{self.name}이(가) 오늘 {self.job}로 일하고 있습니다.")

class Student(Person):
    def __init__(self, name, age, grade, school):
        super().__init__(name, age)
        self.grade = grade
        self.school = school
    
    def introduce(self):
        print(f"저는 {self.school}에서 {self.grade}로 재학중입니다.")

    def change_grade(self):
        print(f"{self.grade}로 변경되었습니다.")

# person1 = Person("Alice", 30)
# person2 = Person("Bob", 25)
# person3 = Person("Danial", 35)

# person1.introduce()
# person2.introduce()
# person3.introduce()

# person1.birthday()
# person2.birthday()
# person3.birthday()

worker1 = Worker("Danial", 35, "연구원")

worker1.introduce()
worker1.birthday()
```

```python
class Animal:
    def sound(self):
        raise NotImplementedError("서브클래스에서 구현해야 한다.")

class Dog(Animal):
    def sound(self):
        return "woof"
    
class Cat(Animal):
    def sound(self):
        return "Meow"
    def eat(self):
        print("오물오물~~")

class Bird(Animal):
    # 오류 1: 부모가 요구한 sound를 구현하지 않으면 에러가 납니다.
    def sound(self):
        return "짹짹"
    
    # 오류 2: 메서드에는 반드시 self가 첫 번째 인자로 들어가야 합니다.
    def play(self):
        return "축구를 합니다"

def main():
    animals = [Dog(), Cat(), Bird()]
    
    for animal in animals:
        print(animal.sound())
    
    # 오류 3: 리스트에서는 .iloc을 쓰지 않고 바로 [index]를 사용합니다.
    print(animals[0].sound())
    print(animals[1].sound())
    print(animals[2].sound())
    print(animals[2].play())

if __name__ == "__main__":
    main()
```

```python
from abc import ABC, abstractmethod

# 추상 클래스: Animal
class Animal(ABC):
    def __init__(self, name):
        self._name = name  # protected attribute

    @abstractmethod
    def make_sound(self):
        """각 동물이 고유의 소리를 내도록 구현"""
        pass

    def get_name(self):
        return self._name

# 자식 클래스: Dog
class Dog(Animal):
    def make_sound(self):
        print("Woof")

# 자식 클래스: Cat
class Cat(Animal):
    def make_sound(self):
        print("Meow")

def main():
    dog = Dog("Buddy")
    cat = Cat("Kitty")

    print(f"{dog.get_name()} makes sound:")
    dog.make_sound()

    print(f"{cat.get_name()} makes sound:")
    cat.make_sound()

if __name__ == "__main__":
    main()
```
