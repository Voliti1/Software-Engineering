## 1.5 변수 스코프와 가시성

```python
# 전역 변수
global_var = "전역 변수"

def outer_function():
    # 함수 스코프 변수
    outer_var = "외부 함수 변수"

    def inner_function():
        # 지역 변수
        local_var = "지역 변수"

        # global 키워드: 전역 변수 수정
        global global_var
        global_var = "수정된 전역 변수"

        # nonlocal 키워드: 상위 함수의 변수 수정
        nonlocal outer_var
        outer_var = "수정된 외부 함수 변수"

        print(f"지역: {local_var}")
        print(f"외부 함수: {outer_var}")
        print(f"전역: {global_var}")

    inner_function()
    print(f"outer_function에서 outer_var: {outer_var}")

def scope_example():
    x = 20  # 지역 (전역 변수를 가림)
    print(f"함수 내부 x: {x}")

# 실행 예제
print(f"수정 전 전역 변수: {global_var}")
outer_function()
print(f"수정 후 전역 변수: {global_var}")

# 스코프 체인 예제
x = 10  # 전역

scope_example()
print(f"전역 x: {x}")
```

```python
x = "전역"

def outer():
    x = "outer"
    def inner():
        x = "inner"
        print(f"inner에서 x: {x}")
    inner()
    print(f"outer에서 x: {x}")

outer()
print(f"전역에서 x: {x}")
```

## 1.6 조건문 vs 예외처리

### 조건문

```python
# 좋은 예 - 사용자 입력은 항상 검증이 필요한 예상 가능한 상황
def search_books(keyword):
    if keyword is None or keyword.strip() == "":
        print("오류: 검색 키워드가 필요합니다.")
        return []

    return perform_search(keyword)

def perform_search(keyword):
    return [f"책 제목: {keyword}"]

print(search_books("파이썬 프로그래밍"))
print(search_books(""))
```

### 예외처리

```python
# 좋은 예 - 파일 읽기 중 발생할 수 있는 다양한 예외들
def read_file(filename):
    try:
        with open(filename, 'r') as f:
            return f.read()
    except FileNotFoundError:
        print(f"파일을 찾을 수 없습니다: {filename}")
    except PermissionError:
        print(f"파일에 접근할 권한이 없습니다: {filename}")
    except Exception as e:
        print(f"파일 읽기 중 오류가 발생했습니다: {e}")
    return None

print(read_file("파일"))
```

```python
# 좋은 예 - 네트워크 오류는 예측할 수 없는 예외 상황
import requests

def fetch_user_data(user_id):
    try:
        response = requests.get(f"https://api.example.com/users/{user_id}")
        return response.json()
    except requests.RequestException as e:
        print(f"네트워크 오류가 발생했습니다: {e}")
        return None


user_data = fetch_user_data(123)
if user_data:
    print(user_data)
else:
    print('데이터를 가져오지 못했습니다.')
```
