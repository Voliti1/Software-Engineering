```python
num = int(input("숫자를 입력하세요 : "))

if num > 0:
    # print(number, "는 양수입니다.")
    print(f"{num}는 양수입니다.")
elif num < 0:
    print(num, "는 음수입니다")
else:
    print("입력한 숫자는 0입니다.")

```

```python
num = int(input("숫자를 입력하세요 : "))
sum = 0

for i in range(1, num + 1):
    sum += i
print(f"1부터 {num}까지의 합은 {sum}입니다")
```
